```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_anfavea = pd.read_excel('../dados/silver/anfavea_producao_veiculos.xlsx', engine='openpyxl')
```

```python
df_anfavea.head()
```

```python
index = 0
lista_blocos = ['AUTOVEÍCULOS TOTAL', 'AUTOMÓVEIS', 'COMERCIAIS LEVES', 'CAMINHÕES', 'ÔNIBUS']
lista_variaveis = ['Emplacamento Total', 'Emplacamento Nacionais', 'Emplacamento Importados', 'Produção', 'Exportação']
for nome in df_anfavea.columns:
    if 'Unnamed: 0' in nome:
        df_anfavea.rename(columns={'Unnamed: 0':'Date'}, inplace=True)
    else:
        if 'Unnamed' not in nome:
            bloco = nome
        df_anfavea.rename(columns={nome:f'{bloco}_{lista_variaveis[index]}'}, inplace=True)
        index += 1
    if index == 5:
        index = 0
df_anfavea.drop(0, axis=0, inplace=True)
```

```python
df_anfavea = df_anfavea[(df_anfavea['Date'] > pd.Timestamp(2012, 12, 31)) & (df_anfavea['Date'] < pd.Timestamp(2026, 3, 1))].copy()
df_anfavea.reset_index(drop=True, inplace=True)
```

```python
df_anfavea.head()
```

```python
colunas_producao = []

for column in df_anfavea.columns:
    if 'Produção' in column:
        colunas_producao.append(column)

producao_total = df_anfavea[colunas_producao].sum(axis=1)

df_anfavea = df_anfavea[['Date']+colunas_producao].copy()
df_anfavea['producao_total'] = producao_total
```
