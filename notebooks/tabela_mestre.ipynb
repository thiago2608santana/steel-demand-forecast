# Variável Alvo

```python
df_variavel_alvo = pd.read_excel(caminho_principal+arquivo_variavel_alvo, engine='openpyxl')
```

```python
df_variavel_alvo.head()
```

```python
df_consumo_aparente = df_variavel_alvo[(df_variavel_alvo['Categoria'] == 'Consumo Aparente / Apparent Consumption (***)') & 
                                        (df_variavel_alvo['Especificação'] == 'Longos / Long Products\n(Inclui Blocos e Tarugos / Included Ingots, Blooms and Billets)')].copy()
df_consumo_aparente.reset_index(drop=True, inplace=True)
df_consumo_aparente.rename(columns={'Valor':'Consumo Aparente', 'Data':'Date'}, inplace=True)
df_consumo_aparente.drop(['Categoria', 'Especificação'], axis=1, inplace=True)
```

```python
df_consumo_aparente.head()
```

```python
df_consumo_aparente.shape
```

# Gerar dataframe com demais variáveis

```python
df = df_consumo_aparente.copy()
for arquivo in lista_arquivos:
    if '.xlsx' in arquivo:
        print(f'Abrindo: {arquivo}...')
        df_temp = pd.read_excel(caminho_principal+arquivo, engine='openpyxl')
        df = df.merge(df_temp, on='Date').copy()
        print(f'-> {arquivo} incluído na tabela mestre!\n')
        #print(f"Data mínima: {df_temp['Date'].min()}")
        #print(f"Data máxima: {df_temp['Date'].max()}")
        print('\n')
```

# Juntar a variável alvo à tabela mestre
